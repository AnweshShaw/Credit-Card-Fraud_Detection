```python
# Training and testing 
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
X = data.drop(['Class'],axis=1)
y = data['Class']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.30,random_state=123)
```

```python
# Random Forest Algorithm
from sklearn.ensemble import RandomForestClassifier
```

```python
rf=RandomForestClassifier()
```

```python
model=rf.fit(X_train,y_train)
```

```python
prediction=model.predict(X_test)
```

```python
rf_accuracy = accuracy_score(y_test,prediction)
```

```python
print("The accuracy of our model using RF classifier = {} %".format(rf_accuracy*100))
```

```python
# Decision Tree Algorithm
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
X2 = data.drop(['Class'],axis=1)
y2 = data['Class']
```

```python
dt = DecisionTreeRegressor()
```

```python
X2_train,X2_test,y2_train,y2_test = train_test_split(X2,y2,test_size=0.3,random_state=123)
```

```python
model2 = dt.fit(X2_train,y2_train)
```

```python
prediction2 = model2.predict(X2_test)
```

```python
dt_accuracy = accuracy_score(y2_test,prediction2)
```

```python
print("The accuracy of our model using DT Regressor is = {} %".format(dt_accuracy*100))
```
